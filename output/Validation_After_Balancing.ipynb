```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("./ttbar/train_data.txt")
```

```python
jet_pt = data[:,0]
jet_eta = data[:,1]
jet_phi = data[:,2]
trk_pt = data[:,3]
trk_eta = data[:,4]
trk_phi = data[:,5]
trk_d0 = data[:,6]
trk_d0_sig = trk_d0 / data[:,7]
trk_z0SinTheta = data[:,8]
trk_z0SinTheta_sig = trk_z0SinTheta / data[:,9]
trk_nPixHits = data[:,10]
trk_nsharedPixHits = data[:,11]
trk_nsplitPixHits = data[:,12]
trk_nSCTHits = data[:,13]
trk_nSCTSharedHits = data[:,14]
trk_nInnermostPixelLayerHits = data[:,15]
trk_nInnermostPixelLayerSharedHits = data[:,16]
trk_nInnermostPixelLayerSplitHits = data[:,17]
trk_nNextToInnermostPixelLayerHits = data[:,18]

label = data[:,19]
ID = data[:,20]

bkg_idx = np.array(np.where(label == 0)).flatten()
sig_idx = np.array(np.where(label == 1)).flatten()

print(len(jet_pt))
```

```python
def plot_hist(var, scale, Bins, Range, Title, Units,log=False):
    plt.title(Title,fontsize = 15)
    plt.hist(var[sig_idx]/scale,bins=Bins,range=Range,histtype=u'step',density=True)
    plt.hist(var[bkg_idx]/scale,bins=Bins,range=Range,histtype=u'step',density=True)
    plt.legend(["sig","bkg"])
    plt.xlabel(Units,loc='right')
    if log==True:
        plt.yscale('log')
    plt.show()
```

```python
delta_R = np.sqrt(np.square(np.subtract(jet_eta,trk_eta))+np.square(np.subtract(jet_phi,trk_phi)))
plot_hist(delta_R,1,20,(0,0.6),"Delta R", "Delta R")
```

```python
plot_hist(jet_pt, 1000, 20, (0,400), "Jet pT", "GeV",log=True)
plot_hist(jet_eta, 1, 40, (-5,5), "Jet Eta", "Eta")
plot_hist(jet_phi, 1, 40, (-5,5), "Jet Phi", "Phi")
plot_hist(trk_pt, 1000, 40, (-20,20), "Track pT/q", "GeV",log=True)
plot_hist(trk_eta, 1, 40, (-5,5), "Track Eta", "Eta")
plot_hist(trk_phi, 1, 40, (-5,5), "Track Phi", "Phi")
plot_hist(trk_d0, 1, 40, (-2,2), "Track d0", "d0",log=True)
plot_hist(trk_d0_sig, 1, 40, (-4,4), "Track d0 Sig", "d0_sig",log=True)
plot_hist(trk_z0SinTheta, 1, 40, (-2,2), "Track z0 SinTheta", "z0SinTheta", log=True)
plot_hist(trk_z0SinTheta_sig, 1, 40, (-4,4), "Track z0 SinTheta Sig", "z0_sig",log=True)
plot_hist(trk_nPixHits, 1, 26, (0,25), "Track nPixHits", "nHits")
plot_hist(trk_nsharedPixHits, 1, 6, (0,5), "Track nSharedPixHits", "nHits",log=True)
plot_hist(trk_nsplitPixHits, 1, 13, (0,12), "trk_nsplitPixHits", "nHits",log=True)
plot_hist(trk_nSCTHits, 1, 20, (0,20), "trk_nSCTHits", "nHits",log=True)
plot_hist(trk_nSCTSharedHits, 1, 7, (0,6), "trk_nSCTSharedHits", "nHits",log=True)
plot_hist(trk_nInnermostPixelLayerHits, 1, 15, (0,14), "trk_nInnermostPixelLayerHits", "nHits",log=True)
plot_hist(trk_nInnermostPixelLayerSharedHits, 1, 5, (0,4), "trk_nInnermostPixelLayerSharedHits", "nHits",log=True)
plot_hist(trk_nInnermostPixelLayerSplitHits, 1, 5, (0,4), "trk_nInnermostPixelLayerSplitHits", "nHits",log=True)
plot_hist(trk_nNextToInnermostPixelLayerHits, 1, 17, (0,16), "trk_nNextToInnermostPixelLayerHits", "nHits",log=True)
```
