```python
import uproot
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = uproot.open("./refined_ttbar.root:Refined_TTree")
```

```python
features = file["features"].array()
trk_label = file["trk_label"].array()
jet_label = file["jet_label"].array()
trk_ID = file["trk_ID"].array()
Event_ID = file["Event_ID"].array()
```

```python
jet_pt = np.array(features[:,0])/1000
jet_eta = np.array(features[:,1])
jet_phi = np.array(features[:,2])
trk_p_over_q = np.array(features[:,3])/1000
trk_eta = np.array(features[:,4])
trk_phi = np.array(features[:,5])
trk_d0 = np.array(features[:,6])
trk_d0_err = np.array(features[:,7])
trk_z0 = np.array(features[:,8])
trk_z0_err = np.array(features[:,9])
trk_nPixHits = np.array(features[:,10])
trk_nSharedPixHits = np.array(features[:,11])
trk_nSplitPixHits = np.array(features[:,12])
trk_nSCTHits = np.array(features[:,13])
trk_nSCTSharedHits = np.array(features[:,14])
trk_nIBLHits = np.array(features[:,15])
trk_nIBLSharedHits = np.array(features[:,16])
trk_nIBLSplitHits = np.array(features[:,17])
trk_nBLHits = np.array(features[:,18])

bkg_idx = np.array(np.where(trk_label == 1)).flatten()
sig_idx = np.array(np.where(trk_label == 0)).flatten()

print("Number of tracks: ", len(trk_label))
```
